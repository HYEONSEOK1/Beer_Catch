```python
# Check the length of images
from glob import glob

# Need to change the extension to 'jpg'
img_list = glob('can/images/*.jpg')
print(len(img_list))
```

```python
# Split dataset to train/valid/test
from sklearn.model_selection import train_test_split

train_img_list, temp_img_list = train_test_split(img_list, test_size=0.3, random_state=11)
val_img_list, test_img_list = train_test_split(temp_img_list, test_size=0.3, random_state=11)

print(len(train_img_list), len(val_img_list), len(test_img_list))

# Create reference files
with open('can/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')

with open('can/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')

with open('can/test.txt', 'w') as f:
    f.write('\n'.join(test_img_list) + '\n')
```
